## Importing dependencies

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

## Extracting html content from koimoi 

```python
html_file = requests.get("https://www.koimoi.com/box-office/bollywoods-100-crore-club/").content
```

## Creating a soup object

```python
soup = BeautifulSoup(html_file, "html.parser")
```

## Extracting Box Office table

```python
boxoffice = soup.find('table', attrs={"id":"tablepress-8"})
```

```python
rows = []
for i in boxoffice.find_all('tr'):
    try:
        rows.append([i.find('td', class_="column-1").get_text(), 
                     i.find('td', class_="column-2").get_text(),
                     i.find('td', class_="column-3").get_text()])
    except:
        pass
```

## Converting the data collected into dataframe

```python
df = pd.DataFrame(rows, columns=['Movie', 'Days to reach 100 crores', 'Total Collection'])
```
